# Patient's Body Temperature throughout the Day

```python
import matplotlib.pyplot as plt


time_of_day = ['8 AM', '10 AM', '12 PM', '2 PM', '4 PM', '6 PM', '8 PM']

temperature_celsius = [98.4, 98.6, 99.0, 99.2, 98.9, 98.7, 98.5]

plt.plot(time_of_day, temperature_celsius, marker='o', color='blue', linestyle='-')

plt.title("Patient's Body Temperature Throughout the Day")
plt.xlabel("Time of Day")
plt.ylabel("Temperature (Â°F)")

plt.grid(True)
plt.show()
```
